```python
import tensorflow as tf
import numpy as np
from matplotlib import pyplot as plt
import cv2
from IPython.display import Image
from tensorflow.keras.preprocessing import image 
from tensorflow.keras.applications import imagenet_utils
```

```python
traffic_img = 'img/traffic_jam.jpeg'
```

```python

Image(filename = 'img/traffic_jam.jpeg', width=224, height=224)
```

# EDA 

```python
load_img = cv2.imread(traffic_img)
```

```python
plt.imshow(load_img)
```

```python
image_resize = cv2.resize(load_img, (224, 224))
```

```python
plt.imshow(image_resize)
```

```python
plt.imshow(cv2.cvtColor(image_resize, cv2.COLOR_BGR2RGB))
```

# Preprocessing of image

```python

img = image.load_img(traffic_img, target_size = (224, 224))
```

```python
plt.imshow(img)
```

```python
resize_img = image.img_to_array(img)
```

```python
final_image = np.expand_dims(resize_img, axis = 0)
```

```python
final_image = tf.keras.applications.mobilenet.preprocess_input(final_image)
```

# Deep Learning model - MobileNet V1

```python
model = tf.keras.applications.mobilenet.MobileNet()
```

```python
predictions = model.predict(final_image)
```

```python
result = imagenet_utils.decode_predictions(predictions)
```

```python
print(result)
```

# Improve results with mobilenet version 2 

```python
model2 = tf.keras.applications.mobilenet_v2.MobileNetV2()
```

```python
predictionsV2 = model2.predict(final_image)
```

```python
result_model2 = imagenet_utils.decode_predictions(predictionsV2)
```

```python
print(result_model2)
```
